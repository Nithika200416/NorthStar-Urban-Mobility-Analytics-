```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base_path = '/content/drive/MyDrive/Colab Notebooks/Raw  Data /'

app_events = pd.read_csv(base_path + 'app_events.csv')
complaints = pd.read_csv(base_path + 'complaints.csv')
customers = pd.read_csv(base_path + 'customers.csv')
deliveries = pd.read_csv(base_path + 'deliveries.csv')
drivers = pd.read_csv(base_path + 'drivers.csv')
hubs = pd.read_csv(base_path + 'hubs.csv')
incidents = pd.read_csv(base_path + 'incidents.csv')
orders = pd.read_csv(base_path + 'orders.csv')
vehicles = pd.read_csv(base_path + 'vehicles.csv')
```

```

print(deliveries.head())
print(orders.head())
print(complaints.head())

print(deliveries.info())
print(orders.info())
```

```


deliveries = deliveries.drop_duplicates()
orders = orders.drop_duplicates()
customers = customers.drop_duplicates()
complaints = complaints.drop_duplicates()


deliveries = deliveries.fillna(0)
orders = orders.fillna(0)
customers = customers.fillna("Unknown")
complaints = complaints.fillna("Unknown")
```

```


if 'order_date' in orders.columns:
    orders['order_date'] = pd.to_datetime(orders['order_date'])

if 'delivery_date' in deliveries.columns:
    deliveries['delivery_date'] = pd.to_datetime(deliveries['delivery_date'])
```

```


df = pd.merge(orders, deliveries, on='order_id', how='left')

df = pd.merge(df, customers, on='customer_id', how='left')

df = pd.merge(df, drivers, on='driver_id', how='left')

df = pd.merge(df, hubs, on='hub_id', how='left')
```

```

save_path = '/content/drive/MyDrive/Colab Notebooks/Processed Data/'

app_events.to_csv(save_path + 'clean_app_events.csv', index=False)

complaints.to_csv(save_path + 'clean_complaints.csv', index=False)

customers.to_csv(save_path + 'clean_customers.csv', index=False)

deliveries.to_csv(save_path + 'clean_deliveries.csv', index=False)

drivers.to_csv(save_path + 'clean_drivers.csv', index=False)

hubs.to_csv(save_path + 'clean_hubs.csv', index=False)

incidents.to_csv(save_path + 'clean_incidents.csv', index=False)

orders.to_csv(save_path + 'clean_orders.csv', index=False)

vehicles.to_csv(save_path + 'clean_vehicles.csv', index=False)


df.to_csv(save_path + 'merged_processed_data.csv', index=False)

print("All cleaned datasets saved successfully")
```

```

if 'status' in df.columns:
    print(df['status'].value_counts())

if 'delay' in df.columns:
    print(df.groupby('hub_id')['delay'].mean())


if 'driver_id' in df.columns and 'delay' in df.columns:
    print(df.groupby('driver_id')['delay'].mean())
```

```

if 'order_id' in complaints.columns:
    df2 = pd.merge(df, complaints, on='order_id', how='left')

    # check which orders have complaints
    df2['has_complaint'] = df2['order_id'].isin(complaints['order_id'])

    print(df2['has_complaint'].value_counts())
```

```

if 'status' in df.columns:

    delivery_status = df['status'].value_counts()

    delivery_status.plot(kind='bar')

    plt.title("Delivery Status Distribution")
    plt.xlabel("Status")
    plt.ylabel("Count")

    plt.show()


if 'hub_id' in df.columns:

    df['hub_id'].value_counts().plot(kind='bar')

    plt.title("Hub Distribution")
    plt.xlabel("Hub ID")
    plt.ylabel("Count")

    plt.show()



if 'complaint_type' in complaints.columns:

    complaint_counts = complaints['complaint_type'].value_counts()

    complaint_counts.plot(kind='bar')

    plt.title("Complaint Types")
    plt.xlabel("Complaint Type")
    plt.ylabel("Count")

    plt.show()



if 'driver_id' in df.columns:

    df['driver_id'].value_counts().head(10).plot(kind='bar')

    plt.title("Top Drivers by Deliveries")
    plt.xlabel("Driver ID")
    plt.ylabel("Number of Deliveries")

    plt.show()
```
