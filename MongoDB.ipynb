```python
import pandas as pd

app_events = pd.read_csv('/content/clean_app_events.csv')

complaints = pd.read_csv('/content/clean_complaints.csv')

customers = pd.read_csv('/content/clean_customers.csv')

deliveries = pd.read_csv('/content/clean_deliveries.csv')

drivers = pd.read_csv('/content/clean_drivers.csv')

hubs = pd.read_csv('/content/clean_hubs.csv')

incidents = pd.read_csv('/content/clean_incidents.csv')

orders = pd.read_csv('/content/clean_orders.csv')

vehicles = pd.read_csv('/content/clean_vehicles.csv')

print("All datasets loaded successfully")
```

```python
app_events_collection = db["app_events"]

complaints_collection = db["complaints"]

customers_collection = db["customers"]

deliveries_collection = db["deliveries"]

drivers_collection = db["drivers"]

hubs_collection = db["hubs"]

incidents_collection = db["incidents"]

orders_collection = db["orders"]

vehicles_collection = db["vehicles"]

print("All collections created successfully")
```

```python
app_events_data = app_events.to_dict(orient='records')

complaints_data = complaints.to_dict(orient='records')

customers_data = customers.to_dict(orient='records')

deliveries_data = deliveries.to_dict(orient='records')

drivers_data = drivers.to_dict(orient='records')

hubs_data = hubs.to_dict(orient='records')

incidents_data = incidents.to_dict(orient='records')

orders_data = orders.to_dict(orient='records')

vehicles_data = vehicles.to_dict(orient='records')

print("All datasets converted to MongoDB documents")
```

```python
app_events_collection.insert_many(app_events_data)

complaints_collection.insert_many(complaints_data)

customers_collection.insert_many(customers_data)

deliveries_collection.insert_many(deliveries_data)

drivers_collection.insert_many(drivers_data)

hubs_collection.insert_many(hubs_data)

incidents_collection.insert_many(incidents_data)

orders_collection.insert_many(orders_data)

vehicles_collection.insert_many(vehicles_data)

print("All datasets inserted successfully into MongoDB Atlas")
```

```python
for data in orders_collection.find().limit(5):
    print(data)
```

```python
for data in orders_collection.find(
    {},
    {"order_id": 1, "customer_id": 1, "service_type": 1, "_id": 0}
).limit(5):

    print(data)
```

```python
high_value_orders = orders_collection.find(
    {
        "order_value": {
            "$gt": 100
        }
    }
)

for data in high_value_orders.limit(5):
    print(data)
```

```python
for data in orders_collection.find().sort("order_value", -1).limit(5):

    print(data)
```

```python
new_order = {
    "order_id": "O5000",
    "customer_id": "C100",
    "service_type": "Priority Delivery"
}

orders_collection.insert_one(new_order)

print("New order inserted successfully")
```

```python
orders_collection.update_one(
    {"order_id": "O5000"},
    {"$set": {"service_type": "Express Delivery"}}
)

print("Order updated successfully")
```

```python
orders_collection.delete_one(
    {"order_id": "O5000"}
)

print("Order deleted successfully")

```

```python
pipeline = [
    {
        "$group": {
            "_id": "$service_type",
            "total_orders": {"$sum": 1}
        }
    }
]

result = orders_collection.aggregate(pipeline)

for data in result:
    print(data)
```

```python
orders_collection.create_index("customer_id")

print("Index created successfully")
```

```python
print(orders_collection.index_information())
```
